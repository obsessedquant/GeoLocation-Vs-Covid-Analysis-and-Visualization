```python
import hvplot.pandas
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('Output/correlations.csv')
df.set_index('Date', inplace=True)
# start_date = pd.to_datetime('2010-01-01')
df.index = pd.to_datetime(df.index)
# df = df[start_date:]
df.tail()
```

```python
df_line = df.hvplot.line(title='10-2 Yield Curve Spread v SP500 Z-scores',
                                   grid=True)
df_line
```

```python
# plt.scatter(df['10-2_Z'],df['SP500_Z'])
# plt.show()
x_values = df['10-2_Z']
y_values = df['SP500_Z']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
df_scatter = df.hvplot.scatter(x='10-2_Z',y='SP500_Z',title='Correlations',
                                   grid=True)
df_scatter
```

```python
correlation = df['10-2_Z'].corr(df['SP500_Z'])
print("Correlation is: ", correlation)
```

```python
correlation_list = []
days_shift_list = []
for x in range(600):
    days_shift = x-300
    df['SP_shift'] = df.SP500_Z.shift(days_shift)
    correlation = df['10-2_Z'].corr(df['SP_shift'])
    correlation_list.append(correlation)
    days_shift_list.append(days_shift)

data_tuples = list(zip(days_shift_list,correlation_list))
correlation_df = pd.DataFrame(data_tuples, columns=['Days_Shifted','Correlation_Coefficient'])
correlation_df.to_csv('correlations.csv')
correlation_df
```
